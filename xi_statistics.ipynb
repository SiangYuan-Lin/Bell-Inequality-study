```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy import genfromtxt
Bxy = genfromtxt('Bxy.csv', delimiter=',')
Byz = genfromtxt('Byz.csv', delimiter=',')
Bzx = genfromtxt('Bzx.csv', delimiter=',')
xip_x = genfromtxt('xip_xy.csv', delimiter = ',')
xip_y = genfromtxt('xip_yz.csv', delimiter = ',')
xip_z = genfromtxt('xip_zx.csv', delimiter = ',')
xim_x = genfromtxt('xim_xy.csv', delimiter = ',')
xim_y = genfromtxt('xim_yz.csv', delimiter = ',')
xim_z = genfromtxt('xim_zx.csv', delimiter = ',')
```

```python
print('Root event count: 132104')
print('Raw event count:',len(Bxy))
print('After tau filter:',len(Bxy[Bxy>-999]))
```

```python
filename = ['Bxy.png','Byz.png','Bzx.png',\
            'xip_x.png','xip_y.png','xip_z.png',\
            'xim_x.png','xip_y.png','xip_z.png']
plt_title = [r'$\langle \mathcal{B}_{xy} \rangle $',r'$ \langle \mathcal{B}_{yz} \rangle$',r'$ \langle \mathcal{B}_{zx} \rangle$',\
             r'$\xi^+_x$',r'$\xi^+_y$',r'$\xi^+_z$',\
             r'$\xi^-_x$',r'$\xi^-_y$',r'$\xi^-_z$']
data = [Bxy[Bxy > -999], Byz[Byz > -999], Bzx[Bzx > -999],\
        xip_x[xip_x > -999], xip_y[xip_y > -999], xip_z[xip_z > -999],\
        xim_x[xim_x > -999], xim_y[xim_y > -999], xim_z[xim_z > -999]]
```

```python
fig, ax = plt.subplots(ncols=3,nrows=3,figsize = (22,20))
for i, ax in enumerate(ax.flat):
    ax.set_title(plt_title[i]+' Distribution')
    ax.set_xlabel(plt_title[i])
    ax.set_ylabel(r'counts')
    ax.hist(data[i][data[i]> -999],bins = 50,edgecolor = 'white')
    mean = np.mean(data[i][data[i]>-999])
    ax.axvline(mean,ls = '--',c = 'red',label = 'mean value = {:.4f}'.format(mean))
    ax.legend()
plt.tight_layout()
plt.savefig("big.png")
plt.show()
```
